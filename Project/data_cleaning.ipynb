```python
import os
import pandas as pd

# read in the data
file_path = os.path.realpath('../Project/Assets/iris.csv')
iris = pd.read_csv(file_path)

iris.head(10)
```

I'm noticing that the class column writes out the full genus-species names which seems unneccessary moving forward seeing as the dataset will continue to only contain flowers of the Iris genus. I will instead rename the class column 'species' and update the values to species-name.

```python
# renaming columns
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris.columns = column_names

# renaming genus-species to species
species_names = {"Iris-setosa": "setosa", "Iris-versicolor": "versicolor", "Iris-virginica": "virginica"}   # create a dictionary of the species names
iris["species"] = iris["species"].map(species_names)                                                        # map the species names to the new names

iris.tail(5)
```

```python
# index=False prevents pandas to write row index into the csv file, will still be added when reading the csv file
iris.to_csv('iris_clean.csv', index=False)  # save the cleaned data to a new csv file
```
